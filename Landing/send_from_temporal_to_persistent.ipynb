Task: move files from temporal to persistent with adding the time in the name.


```python
import os
from google.colab import drive
drive.mount('/content/drive')
```

```python
import shutil
import datetime
from pytz import timezone
import openpyxl

current_time = datetime.datetime.now(timezone('Europe/Madrid')).strftime("%Y-%m-%d %H:%M:%S")
file_name1 = 'worldometer_v1'
new_file_name1 = f'{file_name1}_{current_time}'
file_name2 = 'worldometer_v2'
new_file_name2 = f'{file_name2}_{current_time}'
shutil.move(f'/content/drive/My Drive/Project ADSDB Max_Julian/landing/temporal/{file_name1}.csv', f'/content/drive/My Drive/Project ADSDB Max_Julian/landing/persistent/{new_file_name1}.csv')
shutil.move(f'/content/drive/My Drive/Project ADSDB Max_Julian/landing/temporal/{file_name2}.csv', f'/content/drive/My Drive/Project ADSDB Max_Julian/landing/persistent/{new_file_name2}.csv')

#currently, the event log is in landing. we can change that if necessary
os.chdir('/content/drive/My Drive/Project ADSDB Max_Julian')
book = openpyxl.load_workbook('log.xlsx')
sheet = book.active

row = 2
while sheet[f'A{row}'].value != None:
  row += 1

sheet[f'A{row}'].value = current_time
sheet[f'B{row}'].value = 'File moved from temporal to persistent'
sheet[f'C{row}'].value = f'{file_name1}.csv'
sheet[f'D{row}'].value = f'{new_file_name1}.csv'
row += 1
sheet[f'A{row}'].value = current_time
sheet[f'B{row}'].value = 'File moved from temporal to persistent'
sheet[f'C{row}'].value = f'{file_name2}.csv'
sheet[f'D{row}'].value = f'{new_file_name2}.csv'

book.save('log.xlsx')
```
